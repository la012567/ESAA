```python
# do it pandas
# 데이터 시각화
# 앤스콤 4분할 그래프 살펴보기

import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
# matplotlib 라이브러리 (그래프 그리기)
%matplotlib notebook 
import matplotlib.pyplot as plt

# dataset 열에서 데이터 값이 I인 것만 추출
dataset_1 = anscombe[anscombe['dataset']=='I']

# x, y축 데이터 전달하여 선 그래프 그리기
plt.plot(dataset_1['x'],dataset_1['y'])

# plt.plot(dataset_1['x'],dataset_1['y'],'o') : 점 그래프 그리기
```

```python
# 1. 데이터프레이므의 datset 열 값 추출하여 저장
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']

# 2. 그래프 격자가 위치할 기본 틀 만들기
fig = plt. figure()

# 3. add_subplot(기본 틀 행 크기, 기본 틀 열 크기) 으로 그래프 격자 그리기 
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

# 4. 데이터 전달하여 그래프 그리기
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

# 5. 그래프 격자에 제목 추가하기

axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

# 6. 기본 틀 제목 추가하기
fig.suptitle("Anscombe Data")

# 8. 그래프의 레이아웃 조절하기
fig.tight_layout()
```

```python
# matplotlib 라이브러리 사용


# 1. 기초 그래프 그리기 - 히스토그램, 산점도, 박스
# 1) 데이터 준비

tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))

# 2) 기본 틀 마련하고 그래프 격자 구성하기
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

# 3) 히스토그램 그리기 : 일변량
axes1.hist(tips['total_bill'],bins=10) # hist : 히스토그램, bins : x축 간격
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency') # set_xlabel : x축 제목
axes1.set_ylabel('Total Bill') 
```

```python
# 4) 산점도 그래프 그리기 : 이변량
scatter_plot= plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip']) #scatter :  산점도
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 5) 박스 그래프 : 이산형 + 연속형

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], #boxplot()
               tips[tips['sex'] == 'Male']['tip']],
               labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')

```

```python
# 2. 다변량 그래프 그리기

# 산점도

# 문자열을 정수로 변환하는 함수
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
    
# 함수가 반환한 값 데이터 프레임에 추가하기 (열을 생성해서 ㅈ추가)
tips['sex_color']=tips['sex'].apply(recode_sex) #apply(): sex열에 record_sex 함수 브로드캐스팅

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10, # 테이블당 인원수(size) , s는 점의 사이즈
    c = tips['sex_color'], # c는 점의 색상
    alpha = 0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# seaborn 라이브러리 사용
import seaborn as sns
tips = sns.load_dataset("tips")

ax = plt.subplots() # subplots() : 기본 틀 만들기
ax = sns.distplot(tips['total_bill']) # displot() : 히스토그램 그리기
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],kde=False) # kde = False : 밀집도 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],hist=False) # hist = False : 히스토그램 제외
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],rug=True) # rug=True : 양탄자 그래프 추가
ax.set_title('Total Bill Histogram with Density and Rug plot')
ax.set_xlabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips) # countplot ()
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
# 이변량 그래프

ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips) #regplot (): 회귀선
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg = False) #fit_reg = False : 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x='total_bill',y='tip',data=tips) #jointplot(): 산점도 그래프와 히스토그램 같이
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# hexbin : 육각 그래프 (점이 겹쳐보이지 않음. 데이터 개수가 많아지면 진해짐)
hexbin = sns.jointplot(x='total_bill',y='tip',data=tips, kind="hex") # kind="hex"
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 이차원 밀집도 그리기

ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], # kdeplot():이차원 밀집도
                 data2=tips['tip'],
                 shade = True) #shade = True : 음영효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 바 그래프

ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips) # barplot()
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# 박스 그래프

ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips) # boxplot()
ax.set_title('Boxplot of total Bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')

# 바이올린 그래프

ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips) # violinplot()
ax.set_title('Violin plot of total Bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
# 관계 그래프

fig = sns.pairplot(tips) # pairplot()

# 중복된 정보 제거 : 다른 그래프 채워넣기

pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot) # map_upper : 산점도 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot) # map_lower : 이차원 밀집도
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) #map_diag : 히스토그램과 밀집도
plt.show()
```

```python
# 다변량 그래프

# 바이올린 그래프에 색상 추가
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue= 'sex', data= tips, split= True)
# hue : 색상에 사용할 열 이름
```

```python
# FacetGrid 클래스로 그룹별 그래프 그리기

# 1) 객체 생성하고 그래프 그리기
facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill',rug=True)

# 2) 다변량 그래프
facet = sns.FacetGrid(tips, col='day',hue='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()

# 3) 더 많은 변수를 사용해서 그리기

facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
